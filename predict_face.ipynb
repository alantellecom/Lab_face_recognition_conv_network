```python
from PIL import Image
import cv2
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
model = load_model('.\\facefeatures_new_model.h5')
```

```python
face_cascade = cv2.CascadeClassifier('.\\haarcascade_frontalface_default.xml')
```

```python
def face_extractor_prediction(window):

    faces = face_cascade.detectMultiScale(window, 1.1, 5)
    
    if faces is ():
        return None
    
    for (x,y,w,h) in faces:
        cropped_faces = window[y:y+h, x:x+w]
        face_resize = cv2.resize(cropped_faces, (224, 224)) # formata tamanho para se adequar ao tamnho de fotos de treinamento
        im = Image.fromarray(face_resize, 'RGB') # converte np ndarray(multidimensional array) para imagem
        img_array = np.array(im) # converte imagem para np array
                    #Keras usa tensores 4D, (images x height x width x channel)
                    #então precisamos fazer a converssão: 224x224x3 into 1x224x224x3 
        img_array = np.expand_dims(img_array, axis=0)
        pred = model.predict(img_array) # realiza a predição
        print(pred)
        if(pred[0][0]>0.5): # probabilidade acima de 0.5, classifica como Tifani, conforme definição do modelo
            name='Tifani'
        else:
            name='Outros'
        k=10    
        cv2.rectangle(window,(x,y),(x+w,y+h),(0,255,255),2) # cria retangulo em torno das cordenadas da face detectada
        cv2.putText(window,name, (x-k, y-k), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2) # coloca texto da classificação sobre face detectada
        cv2.imshow('Video', window) # Exibe
#--------------------------------------------------------------------------------------------------------------------------
video_capture = cv2.VideoCapture(0)

while True:
    _, window = video_capture.read()

    face_extractor_prediction(window)
        
    if cv2.waitKey(1) == 27: # 27 = tecla 'esc'
        break
               
video_capture.release()
cv2.destroyAllWindows()
```
